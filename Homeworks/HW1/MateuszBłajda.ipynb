## Importing Libraries 

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

from sklearn.model_selection import train_test_split
import numpy as np
```

## Loading Dataset
loading brain stroke dataset

```python
dataset_path = '../../../brain_stroke.csv'
df = pd.read_csv(dataset_path)
df.head()
```

## Data preprocessing

Changing text values to binary or one hot encoding

```python
df['gender'] = np.where(df['gender'] == 'Female', 1, 0)
df['ever_married'] = np.where(df['ever_married'] == 'Yes', 1, 0)
df['Residence_type'] = np.where(df['Residence_type'] == 'Urban', 1, 0)
w = pd.get_dummies(df['work_type'], prefix='work_type')
df = df.drop('work_type', axis=1).join(w)
s = pd.get_dummies(df['smoking_status'], prefix='smoking')
df = df.drop('smoking_status', axis=1).join(s)
df.head()
```

## Preparing the dataset
spliting into train and test dataset (randomly at 8:2 ratio)

```python
train, test = train_test_split(df, test_size= 0.2)
trainX = train.drop('stroke', axis=1)
trainY = train['stroke']

testX = test.drop('stroke', axis=1)
testY = test['stroke']
```

## Training the models
I trained 3 different models on this dataset:
* Logistic regression
* Random Forest
* AdaBoost

```python
model1 = LogisticRegression(max_iter=1000, random_state=42).fit(trainX, trainY)
score1 = model1.score(testX, testY)
```

```python
model2 = RandomForestClassifier(n_estimators=8, random_state=42).fit(trainX, trainY)
score2 = model2.score(testX, testY)
```

```python
model3 = AdaBoostClassifier(n_estimators=8, random_state=42).fit(trainX, trainY)
score3 = model3.score(testX, testY)
```

```python
print(f"Logistic regression score: {score1}")
print(f"Random Forest score: {score2}")
print(f"AdaBoost score: {score3}")
```

## Results:
Every model performed well, Random forest was a little bit worse, AdaBoost and Logistic regression both had literally the same score
